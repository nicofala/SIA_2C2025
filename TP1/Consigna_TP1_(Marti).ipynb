 **Trabjaremos con modelos lineales**

Son aquellos que tienen la forma:

$\hat{y} =\theta_0 +\theta_1 x_1 + \theta_2 x_2 + .. +\theta_n x_n$

o, vectorialmente

$\hat{y} = \vec{\theta} \cdot \vec{x}$

donde $\hat{y}$ es la predicción,
$n$ es el número de _features_ o variables predictoras, $x_i$ es el i-ésimo feature y $\vec{\theta}$ es el vector de parámetros o _pesos_ del modelo (lo que entrenamos).

**Modelos lineales regularizados**

En los modelos lineales simples, los parámetros $\theta_i$ se ajustan minimizando el error cuadrático medio entre las predicciones y los valores reales. Sin embargo, cuando hay **muchas variables**, correlaciones fuertes entre ellas o riesgo de *overfitting*, se utilizan **términos de regularización** que penalizan la magnitud de los coeficientes, mejorando así la capacidad de generalización del modelo.

Algunos ejemplos:

- **Ridge Regression** (*L2 regularization*):  

  $\text{Loss} = \text{MSE} + \alpha \sum_{i=1}^n \theta_i^2$

  Tiende a mantener todos los coeficientes pequeños pero distintos de cero.

- **Lasso Regression** (*L1 regularization*):  

  $\text{Loss} = \text{MSE} + \alpha \sum_{i=1}^n |\theta_i|$

  Puede forzar coeficientes exactamente a cero, funcionando también como método de selección de variables.

- **Elastic Net**: combina L1 y L2, balanceando sus ventajas.


**Modelos lineales con *features* polinomiales**

Los modelos lineales pueden extenderse para capturar relaciones **no lineales** entre las variables de entrada y la variable objetivo mediante la creación de nuevas variables que son **combinaciones polinomiales** de las originales.

Por ejemplo, con dos variables $x_1$ y $x_2$ y un polinomio de grado 2, el modelo considera:  
$
\hat{y} = \theta_0 + \theta_1 x_1 + \theta_2 x_2 + \theta_3 x_1^2 + \theta_4 x_2^2 + \theta_5 x_1 x_2
$  

Aunque la relación entre las *features* y $\hat{y}$ puede ser no lineal, el modelo sigue siendo **lineal en los parámetros** $\theta_i$, por lo que conserva las propiedades y métodos de ajuste de los modelos lineales clásicos.

Este enfoque, combinado con regularización, permite capturar patrones complejos sin perder el control sobre el sobreajuste.

1. Reglas de negocio (ej. rooms ≤ 10, bathrooms ≤ 6)

Por qué: en datos inmobiliarios hay cosas que sabemos de antemano que son imposibles o irreales.

Ejemplo: un departamento con 50 baños no tiene sentido → aunque estadísticamente no sea “outlier”, lo eliminamos.

Ventaja: garantiza que los datos tengan coherencia con la realidad del dominio.

Cuándo usarlo: siempre que sepas rangos lógicos de tu variable.

🔹 2. Regla de IQR (Tukey)

Por qué: mide la dispersión real de la mayoría de los datos.

Se basa en el rango intercuartílico (Q3–Q1), que captura el 50% central de la distribución.

Los valores fuera de Q1 − 1.5·IQR y Q3 + 1.5·IQR suelen ser considerados atípicos.

Ventaja: es robusto a distribuciones sesgadas (ejemplo: precios, superficies).

Cuándo usarlo: cuando querés eliminar valores que están demasiado lejos del “cuerpo” de la distribución.

🔹 3. Percentiles (1%–99%)

Por qué: en datasets grandes, siempre aparecen colas muy largas (ejemplo: casas gigantes, precios millonarios).

Cortar en percentiles evita que estos extremos dominen la escala de análisis o distorsionen modelos.

Ventaja: simple, flexible y no depende de la forma de la distribución.

Cuándo usarlo: cuando querés garantizar que eliminás los extremos más raros aunque no los detecte IQR.

🔹 4. Usar el rango más estricto entre IQR y percentiles

Por qué:

A veces el IQR da un rango demasiado amplio y deja pasar outliers.

Otras veces los percentiles cortan de más y eliminan datos válidos.

Combinarlos y quedarte con el más restrictivo te da un balance:

Protegés los datos “centrales” con IQR.

Y al mismo tiempo controlás los extremos con percentiles.

✅ En resumen:

Reglas de negocio = evitar absurdos.

IQR = capturar la dispersión real de los datos.

Percentiles = controlar colas largas o distribuciones muy sesgadas.

Combinación = limpieza robusta, sin perder demasiada data válida.

```python
# -----------------------------
# Limpieza de outliers combinada
# -----------------------------

def limpiar_outliers(df):
    df_clean = df.copy()

    # --- Variables de conteo ---
    reglas_conteo = {
        'rooms': (1, 10),       # mínimo 1, máximo 10
        'bedrooms': (0, 8),     # puede haber 0 (monoambiente), límite razonable 8
        'bathrooms': (0, 6)     # límite razonable 6
    }

    for col, (min_val, max_val) in reglas_conteo.items():
        df_clean = df_clean[(df_clean[col] >= min_val) & (df_clean[col] <= max_val)]

    # --- Variables continuas ---
    continuas = ['surface_total', 'surface_covered', 'price']

    for col in continuas:
        # IQR
        Q1 = df_clean[col].quantile(0.25)
        Q3 = df_clean[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_iqr = Q1 - 1.5 * IQR
        upper_iqr = Q3 + 1.5 * IQR

        # Percentiles
        lower_pct = df_clean[col].quantile(0.01)
        upper_pct = df_clean[col].quantile(0.99)

        # Usamos el rango más estricto entre IQR y percentiles
        lower = max(lower_iqr, lower_pct)
        upper = min(upper_iqr, upper_pct)

        df_clean = df_clean[(df_clean[col] >= lower) & (df_clean[col] <= upper)]

    return df_clean

# Aplicar limpieza
df_clean = limpiar_outliers(df)
print(f"Antes: {df.shape[0]} filas")
print(f"Después: {df_clean.shape[0]} filas")
```
